# Question 2

```python
"""
As part of your role in a data-driven project, you need to take input from the user for their age, weight, and height. However, there's an issue:
the user has entered their age as a negative number. You need to correct the age of the user, and after correcting the age, calculate their Basal
Metabolic Rate (BMR) using the Harris-Benedict equation for men. Ensure that the final answer of BMR is an integer.  

Basal Metabolic Rate (BMR):
Formula (Harris-Benedict equation for men):
BMR = 88.362 + (13.397 * weight in kg) + (4.799 * height in cm) - (5.677 * age in years)
"""
age = int(input("Enter your age"))
age = abs(age)
weight = int(input("Enter your weight in kg:"))
height = int(input("Enter your height in cm:"))
BMR = 88.362 + (13.397 * weight) + (4.799 * height) - (5.677 * age)
print(BMR)
```
