# Functions

# Why we need functions in python???
We need functions in python for few reasons:
- Reuse of code: Allow us to write a block of code once and use it multiple time
- Organization of code: Help to organize code in small and manageable piece

# Types of functions in python
- user defined functions
- Built in functions
- lambda functions
- recursion functions

# Built in Functions

# 1. Eval function

```python
a = 5
b = 10
expression = 'a+b*2'
result = eval(expression)
print(result)
```

```python
a = 5
b = 10
expression = 'a+b*2/36*6232+2323/2323*6632'
result = eval(expression)
print(result)
```

# 2. ABS Function (Absolute function)

```python
#Always change the value into absolute format means change the value from negative to positive
a = -10
print (abs(b))
```

# 3. sum function

```python
# used to sum the arrays values
xyz = [1,2,3,4,5]
total = sum(xyz)
print(total)
```

# 4. Pow function

```python
# used to find power of any number

result = pow(2,12)
print(result)
```

# 5. Input functions

```python
# used to take input from user
# It always take string type data

a = input("Please enter your age:")
print(a)
print(type(a))
```

```python
#Convert into required Data type

a = int(input("Please enter your age:"))
print(a)
print(type(a))
```

```python
# int to str

a = str(5+5)
print (a)
print(type(a))
```

```python
a = input("Please enter your age:")
print(a)
print(type(a))
b = float(a)
print(b)
print(type(b))
```
