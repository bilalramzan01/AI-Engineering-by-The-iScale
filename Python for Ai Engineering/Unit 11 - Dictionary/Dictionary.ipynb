# Dictionary
- container use to store key value pairs
- enclosed in {} Curly braces separted by , and key value separated by collons:
- Dictionary Items are ordered.
- Keys Duplicates are not allowed.
- Dict is mutuable
- 

```python
a = {
    "Name" : "Bilal" ,
    "ID" : 12,
    "Age": 23,
}
print(a)
print(type(a))
```

```python
a = {
    "Name" : "Bilal" ,
    "ID" : 12,
    "Age": 23,
    "Age": 25,
}
print(a["Age"])
print(len(a))
```

# Dict Constructor

```python
a = dict(name="Bilal", id=23)
print(a)
print(type(a))
```

# Mutuable

```python
# Replace value
a = {
    "Name" : "Bilal" ,
    "ID" : 12,
    "Age": 23,
    "Age": 25,
}
a["Name"] = "Afzal"
print(a)
```

```python
# Add new key value
a = {
    "Name" : "Bilal" ,
    "ID" : 12,
    "Age": 23,
    "Age": 25,
}
a.update({"Mail id": "bilalramzan.cs@gmail.com"})
print(a)
```

```python
# remove any value
a = {
    "Name" : "Bilal" ,
    "ID" : 12,
    "Age": 23,
    "Age": 25,
}
a.pop("ID")
print(a)
```

```python
# clear all the dic values

a = {
    "Name" : "Bilal" ,
    "ID" : 12,
    "Age": 23,
    "Age": 25,
}
a.clear()
print(a)
```
