# while loop
In python, a while loop is used to execute a block of statements repeatedly until a given condition is satisfied. And when the condition becomes false, the line immediately after the loop in the program is executed.


```python
i = 1
n = 5 #Can run infinite times
while i<=n:
    print(i)
    i=i+1
```

# For loop
The for loop in Python is indeed used to iterate over sequences such as lists, tuples, strings, dictionaries, and ranges. It allows you to execute a block of code for each item in the sequence.

```python
#Print the characters of the words
a = "datascience"
for i in a:
    print(i)
```

```python
# print numbers from 1 to 5

for i in range(6):
    print(i)
```

```python
for i in range(2,6):
    print(i)
```

```python
a = [10,20,30,40,50]
for i in a:
    print(i)
```

```python
a = (10,20,30,40,50)
for i in a:
    print(i)
```

# Break and Continue Statment

```python
#Break statmentS
for i in range(5):
    if i==3:
        break
    print (i)
```

```python
for i in range(10):
    if i==3:
        continue
    print (i)
print("I am Bilal Ramzan")
```
