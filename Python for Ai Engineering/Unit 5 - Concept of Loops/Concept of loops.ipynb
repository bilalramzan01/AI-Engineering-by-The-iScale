# while loop
In python, a while loop is used to execute a block of statements repeatedly until a given condition is satisfied. And when the condition becomes false, the line immediately after the loop in the program is executed.


```python
i = 1
n = 5
while i<=n:
    print(i)
    i=i+1
```

# For loop
The for loop in Python is indeed used to iterate over sequences such as lists, tuples, strings, dictionaries, and ranges. It allows you to execute a block of code for each item in the sequence.
