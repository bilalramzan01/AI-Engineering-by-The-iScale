# Control flow statment
Control flow statements in Python are instructions that determine the order in which code is executed based on certain conditions or loops.
They allow you to control the flow of execution within a program, deciding which blocks of code to execute and when to execute them.
there are three types of CFS
1. Conditional statments
2. Transfer statments
3. Iterative statments


# Conditional Statments 
- If statment
- if-else stament
- if-elif-else statment
- nested if-else statment

# If statment

```python
i = 20
if (i>10):
    print("10 is less then 20")
print("I am outside the block")
```

```python
i = int(input("Enter your age:"))
if (i>18):
    print("You are qualified")
print("I am outside the block")
```

# If-else statment

```python
age = int(input("enter your age"))

if age>18:
    print("Your are eligibale for vote:")
else:
    print("Not eligiable")
```

# if-elif-else statment

```python
# Lahore = Heart of Pakistan
# Karachi = City of lights
# Islamabad = Capital City
# Multan = City of Mangos
# Quetta = Biggest city

city = input("Enter your City name")

if city== 'Lahore':
    print("Heart of Pakistan")
elif city== 'Karachi':
     print("City of lights")
elif city== 'Islamabad':
     print("Capital City")
elif city== 'Multan':
     print("City of Mangos")
elif city== 'Quetta':
    print("Biggest city")
else:
    print("None of them")
```
