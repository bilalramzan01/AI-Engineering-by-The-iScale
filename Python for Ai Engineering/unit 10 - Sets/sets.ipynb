# Sets in python
- Container use to store unique data values.
- Value are unordered.
- Duplicates not allowed
- 1 and True are considered one time and considered same value
- nesting is not allowed

```python
a = {20,30,10,40,50}
print(a)
print(type(a))
```

# Duplicates not allowed

```python
a = {"Cat","Dog","Elephant","Jiraph",50,50,4,50,1,True}
print(a)
print(type(a))
print(len(a))
```

# Set constructor

```python
a = set((1,2,3,4,5))
print(a)
print(type(a))
```

# Empty set

```python
a = set()
print(type(a))
```

# ADD and Remove value


```python
a = set((1,2,3,4,5))
a.add("Bilal")
print(a)
print(type(a))
```

```python
a = set((1,2,3,4,5))
a.remove(5)
print(a)
print(type(a))
```

```python
a = set((1,2,3,4,5))
a.clear()
print(a)
print(type(a))
```

# Join two sets

```python
a = set((1,2,3,4,5))
b = set((6,7,8,9))
print(a.union(b))
```
