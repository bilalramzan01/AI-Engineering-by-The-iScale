# List
- Container used to store a All type of data in a single variable.
- it used square brackets [] and separated by comas.
- It allows duplicate values
- It has indexing
- List is Mutuable


```python
fruits = ["Apple","Banana","Cherry", "Grapes"]
print(fruits)
print(type(fruits))
```

```python
a = [12,23.45,"Cherry", "Grapes"]
print(a)
print(type(a))
```

```python
a = [12,23.45,"Cherry", "Grapes",12,12, True, False]
print(a)
print(type(a))
```

```python
a = [12,23.45,"Cherry", "Grapes",12,12, True, False]
print(a)
print(type(a))
print(len(a))
```

# How to access items

```python
a = [12,23.45,"Cherry", "Grapes",12,12, True, False]
print(a[2])
print(a[-2])
```

```python
# String Slicing

a = [10,20,30,40,50,60, 70, 80]
print(a[2:6])
```

```python
# Gap of 2
a = [10,20,30,40,50,60, 70, 80]
print(a[2:6:2])
```

# Reverse of list

```python
a = [10,20,30,40,50,60, 70, 80]
print(a[::-1])
```

# List is Mutuable

```python
a = [10,20,30,40,50,60, 70, 80]
a[1]=5
print(a)
```

# Append and Remove

```python
a = [10,20,30,40,50,60, 70, 80]
a.append(90)
print(a)
```

```python
a = [10,20,30,40,50,60, 70, 80]
a.remove(70)
print(a)
```

# Joining 2 lists

```python
a = [10,20,30]
b = [40,50,60]
print(a+b)
```

```python
a = [10,20,30]
b = [40,50,60]
a.reverse()
print(a)
```
