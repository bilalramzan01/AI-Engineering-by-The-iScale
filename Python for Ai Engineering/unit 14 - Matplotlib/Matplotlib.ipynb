# Matplotlib: Visualization with Python
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.
Matplotlib is a Python library for creating static, interactive and animated visualizations from data. It provides flexible and customizable plotting functions that help in understanding data patterns, trends and relationships effectively.

```python
import matplotlib.pyplot as plt
```

# Linear Graph

```python
x = [2,3,4,5,6]
y = [10,20,30,40,50]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

```python
x = [10.4,3.1,8.7,9,6]
y = [10,20,30,40,50]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

```python
x = [10.4,3.1,8.7,9,6]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.plot(y, marker = 'o')
plt.show()
```

# Bar Graph

```python
x = [10.4,3.1,8.7,9,6]
y = [10,20,30,40,50]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
c = ['r','b','y','g']
plt.bar(x,y,color = c)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array(['A', 'B', 'C','D'])
y = np.array([3,5,6,8])
plt.barh(x,y)
plt.show()
```

```python
x = np.array(['A', 'B', 'C','D'])
y = np.array([3,5,6,8])
# c = ['r']
plt.barh(x,y, color = 'pink')
plt.show()
```

```python
x = np.array(['A', 'B', 'C','D'])
y = np.array([3,5,6,8])
plt.barh(x,y,height=0.1)
plt.show()
```
