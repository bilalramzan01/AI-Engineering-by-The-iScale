# Variables in Python 
Variables in python are like a containers that holds data values. when you assign a value to the container you are putting a value to that container.
they are created when you assign a value to them.
 

# Rules for python Variables
- Always start with a letter or underscore value
- Cannot start with a number
- variable contain alphnumeric and underscore values
- variables are case sensitive
- variable cannot be of any python keywords

```python
abc = "Hello"
print(abc)
```

```python
_abc = "Hello"
print(_abc)
```

# Data Types

```python
# Integers data types. Data having numbers only is an integers
x = 2
print(x)
print(type(x))
```

```python
# Float data types. Data having numbers in points
y = 2.5
print(y)
print(type(y))
```

```python
# String data type. Data having sequence of characters in it and enclosed in single or double quotes
z = "Bilal"
print(z)
print(type(z))
```

```python
# complex data. data having ayota in it
a = 2+3j
print(a)
print(type(a))
```

# Operators in Python

```python
# Addition: Addition operator +

a = 3
b = 5
print(a+b)
```

```python
# Subtraction: - used for subtraction operator

a = 8
b = 5
print(a-b)
```

```python
# Multiplication operator: * is used for multiplication operator

a = 8
b = 5
print(a*b)
```

```python
# Division operator: / is used for division operator

a = 8
b = 5
print(a/b)
```

```python
# Modulus operator. % is used for modulues opertor
a = 8
b = 5
print(a%b)
```

```python
# Exponent operator: ** used for exponent operator

a = 8
print(a**6)
```
